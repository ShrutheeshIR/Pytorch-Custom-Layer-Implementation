```python
import torch
import torch.nn as nn
import torch.nn.functional as F
    from torch impautogradgrad
import torchvision
import torchvision.transforms as transforms
```

```python
transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])

trainset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=4,
                                          shuffle=True, num_workers=2)

testset = torchvision.datasets.CIFAR10(root='./data', train=False,
                                       download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=4,
                                         shuffle=False, num_workers=2)

classes = ('plane', 'car', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

```python
class Net(nn.Module):
    def __init__(self):
        
        super().__init__()
        
        self.act_max = autograd.Variable(torch.rand(1), requires_grad=True)
        self.conv1 = nn.Conv2d(3, 32, kernel_size=5)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=5)
        self.pool = nn.MaxPool2d(kernel_size = 5, return_indices=True)
        self.deconv1 = nn.ConvTranspose2d(64, 32, kernel_size=5)
        self.maxunpool1 = nn.MaxUnpool2d(kernel_size=5)
        self.deconv2 = nn.ConvTranspose2d(32, 3, kernel_size=5)
        

    def forward(self, input):
        conv1 = self.conv1(input)
        
#         print('Conv1')
#         print(conv1.shape)
        pool1,indices1 = self.pool(conv1)
        size1=conv1.size()
#         print('Pool1')
#         print(pool1.shape)
#         print(indices1.shape)
#         pool1[pool1 < self.act_max] = 0
#         tensor123 = torch.zeros((1,1), dtype=torch.int8)
#         print(tensor123)
#         print(self.act_max)
        x=torch.zeros((1,1), requires_grad=True)
#         print(x)
#         print(self.act_max)
        pool1 = torch.where(pool1 < self.act_max, self.act_max, pool1)
#         pool1[pool1 < self.act_max] = torch.tensor([0], dtype=torch.long)
        conv2 = self.conv2(pool1)
#         print('Conv2')
#         print(conv2.shape)
        deconv1=self.deconv1(conv2)
#         print('Deconv1')
#         print(deconv1.shape)
#         print(indices1.shape)
#         print(size1)
        maxunpool1=self.maxunpool1(deconv1, indices1,size1)
#         print('Maxunpool1')
#         print(maxunpool1.shape)
        deconv2=self.deconv2(maxunpool1)
#         print('deconv2')
#         print(deconv2.shape)
        return deconv2
```

```python
model=Net()
print(model)
```

```python
print(model.parameters)
```

```python
import torch.optim as optim

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3,
                             weight_decay=1e-5)
```

```python
for epoch in range(20):  # loop over the dataset multiple times

    running_loss = 0.0
    for i, data in enumerate(trainloader, 0):
        # get the inputs
        inputs, labels = data
#         print(inputs.shape)
        

        # zero the parameter gradients
        optimizer.zero_grad()
#         print('***')

        # forward + backward + optimize
        outputs = model(inputs)
#         print('$$$')
#         print(outputs.shape)
        loss = criterion(outputs, inputs)
        loss.backward()
        optimizer.step()
#         print('*****asdas')
#         print(model.act_max._grad)
        model.act_max.data = model.act_max.data - 0.001 * model.act_max._grad.data
        # print statistics
        running_loss += loss.item()
        if i % 100 == 99:    # print every 100 mini-batches
            print('[%d, %5d] loss: %.3f ' %
                  (epoch + 1, i + 1, running_loss / 2000))
            running_loss = 0.0
            print(model.act_max.data.item())

print('Finished Training')
```
